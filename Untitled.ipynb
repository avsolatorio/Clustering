```python
import cPickle
filename = 'Educ_Industry_data.list'
with open(filename, 'r') as f:
    raw_data = cPickle.load(f)
```

```python
education, work_experience = zip(*raw_data)
```

```python
schools = map(lambda x: x['education']['school'], filter(lambda x: 'school' in x['education'], education))
```

```python
import main
for i in xrange(len(schools)):
    s = ''
    for ch in schools[i]:
        try:
            ch = str(ch)
            s += ch
        except:
            pass
    schools[i] = s
```

```python
clean_schools = main.clean(schools, 'industry')
```

```python
import scluster

%time sc = scluster.lattice_spanning(list(set(clean_schools)), min_common = 8)
```
